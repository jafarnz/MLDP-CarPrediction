```python
## Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly
```

```python
## Read data
FILE_PATH= "corr.xlsx"

for i in range(1, 10):
    df = pd.read_excel(FILE_PATH, sheet_name=f'Sheet{i}')
    col_x = 'x'
    col_y = 'y'
    x = df[col_x]
    y = df[col_y]

    # Print summary statistics
    print(df.describe())

    # Calculate means
    mean_x = np.mean(x)
    mean_y = np.mean(y)

    # Pearson correlation coefficient (r)
    numerator = np.sum((x - mean_x) * (y - mean_y))
    denominator = np.sqrt(np.sum((x - mean_x)**2) * np.sum((y - mean_y)**2))
    r = numerator / denominator
    # r = df[col_x].corr(df[col_y])

    # Coefficient of determination (r^2) for 2 variables
    r_squared = r ** 2

    print(f"Pearson correlation coefficient (r): {r}")
    print(f"Coefficient of determination (r^2): {r_squared}")

    df.plot(x=col_x, y=col_y, kind='scatter')
    plt.show()
```
