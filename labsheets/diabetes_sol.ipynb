# 3. Data Preparation

## 3.1 Data Cleaning

```python
## We will just run the code without data cleaning for demonstration purpose

col_irrelevant = [col_y, 'GenHlth', 'MentHlth', 'PhysHlth', 'Education', 'Income']
```

## 3.2 Handling Categorical Data (One-Hot Encoding)

```python
## Check datatypes for each column
col_categorical = df.select_dtypes(include=['object']).columns ## Select all categorical columns

## Print data about categorical column
'''
nunique(): Return total number of unique values in column
unique(): Return unique values in column
'''
for col in df[col_categorical]:
    print(f'{col} ({df[col].nunique()}): {df[col].unique()}') 
```

```python
## Split data into features (X) and target (y)
y = df[col_y] ## Select target column 

# col_irrelevant = [col_y]

X = df.drop(col_irrelevant, axis=1) ## Select feature columns 

## One-Hot Encoding
X = pd.get_dummies(X, 
               drop_first=True ## Remove redundant information
               )
X
```

## 3.3 Train-Test Split

```python
from sklearn.model_selection import train_test_split

## Split data into train set and test set
test_size = 0.3 
random_state = 2025 ## For reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=test_size, 
                                                    random_state=random_state)
```

### ü§ì Back to slides üë®üèª‚Äçüè´

# 4. Modelling

### 4.2 Train Model

```python
from sklearn.tree import DecisionTreeClassifier

## Initialise model
dt = DecisionTreeClassifier() ## Decision Tree Regressor model
dt.fit(X_train, y_train) ## Train model

## Evaluate model
from sklearn.metrics import accuracy_score
y_pred_dt = dt.predict(X_test)
print("Decision Tree Accuracy: ", accuracy_score(y_test, y_pred_dt))

## Save model
import joblib
joblib.dump(dt, 'diabetes_dt_model.pkl') ## Save Decision Tree model
```

```python
## Initialise model
from sklearn.ensemble import GradientBoostingClassifier
gbt = GradientBoostingClassifier()
gbt.fit(X_train, y_train) ## Train model

## Evaluate model
from sklearn.metrics import accuracy_score
y_pred_gbt = dt.predict(X_test)
print("Gradient Boosted Tree Accuracy: ", accuracy_score(y_test, y_pred_gbt))

## Save model
import joblib
joblib.dump(gbt, 'diabetes_gbt_model.pkl') ## Save Gradient Boosted Tree model
```

```python
## Initialise model
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train) ## Train model

## Evaluate model
from sklearn.metrics import accuracy_score
y_pred_rf = rf.predict(X_test)
print("Random Forest Accuracy: ", accuracy_score(y_test, y_pred_rf))

## Save model
import joblib
joblib.dump(gbt, 'diabetes_rf_model.pkl') ## Save Gradient Boosted Tree model
```

```python
## Define parameter grid
param_grid_rf = {
    'n_estimators': [10, 50], ## Number of trees in the forest
    'criterion': ['gini', 'entropy'], ## Measure quality of a split
    'max_depth': [1, 2, 3], ## Maximum depth of tree
    'min_samples_split': [2, 5], ## Minimum number of samples required to split a node
    'max_features': ['sqrt', 'log2'] ## Number of features to consider at each split
    }

## Initialise model
from sklearn.model_selection import GridSearchCV
gs_rf = GridSearchCV(estimator = rf,
                     param_grid = param_grid_rf,
                     cv = 5, ## Number of cross-validation folds
                     scoring = 'accuracy', ## Scoring metric
                     )

gs_rf.fit(X_train, y_train)

## Obtain best parameters and best estimator
best_gs_params = gs_rf.best_params_
best_gs_rf = gs_rf.best_estimator_
print("Best Parameters: ", best_gs_params)
print("Best Estimator: ", best_gs_rf)

## Evaluate model
y_pred_gs_rf = best_gs_rf.predict(X_test) ## Prediction using model 3
print("Grid Search Random Forest Classifier accuracy: ", accuracy_score(y_test, y_pred_gs_rf))

# Save model
import joblib
joblib.dump(best_gs_rf, 'diabetes_best_gs_rf_model.pkl') ## Save Random Forest model
```

```python
## Define parameter distribution
from scipy.stats import randint, uniform

param_dist_rf = {
    'n_estimators': randint(1, 5),  ## Number of trees in the forest
    'criterion': ['gini', 'entropy'], ## Measure quality of a split
    'max_depth': randint(1, 3),   ## Maximum depth of tree
    'min_samples_split': randint(2, 5),  ## Minimum number of samples required to split a node
    'max_features': ['sqrt', 'log2'],  ## Number of features to consider at each split
    'max_samples': uniform(0.6, 0.4)  ## range [0.6, 1.0) ## Fraction of samples to use for training each tree
}

from sklearn.model_selection import RandomizedSearchCV
rs_rf = RandomizedSearchCV(
    estimator = rf, 
    param_distributions = param_dist_rf,
    cv = 5, ## Number of cross-validation folds
    scoring = 'accuracy', ## Scoring metric
    n_iter = 10, ## Number of parameter settings sampled
    n_jobs = -1 ## -1: Use all available cores
    )

rs_rf.fit(X_train, y_train)

## Obtain best parameters and best estimator
best_rs_params = rs_rf.best_params_
best_rs_rf = rs_rf.best_estimator_
print("Best Parameters: ", best_rs_params)
print("Best Estimator: ", best_rs_rf)

## Evaluate model
y_pred_rs_rf = best_rs_rf.predict(X_test) ## Prediction using model 3
print("Randomized Search Random Forest Classifier accuracy: ", accuracy_score(y_test, y_pred_rs_rf))

# Save model
import joblib
joblib.dump(best_rs_rf, 'diabetes_best_rs_rf_model.pkl') ## Save Random Forest model
```
